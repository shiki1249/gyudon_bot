```python
def convert_label(label):
    l = []
    for y in label:
        tmp = []
        for i in range(10):
            tmp.append(1) if i == y else tmp.append(0)
        l.append(tmp)
    return l
```

```python
import random
tmp = [random.randint(0,100) for i in range(0,100)]
convert_label(tmp)
```
